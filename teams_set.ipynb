```python
import pandas as pd
```

```python
teams = pd.read_csv("teams.csv")
```

```python
teams = teams[["team", "country", "year", "athletes", "age", "prev_medals", "medals"]]
```

```python
numeric_teams = teams[["year", "athletes", "age", "prev_medals", "medals"]]
```

```python
print("Checking for non-numeric values in the numerical columns:")
for column in numeric_teams.columns:
    if not pd.api.types.is_numeric_dtype(numeric_teams[column]):
        print(f"Non-numeric values found in column: {column}")
        print(numeric_teams[column].unique())

correlation_matrix = numeric_teams.corr()
```

```python
print("Correlation Matrix:")
print(correlation_matrix)

if "medals" in correlation_matrix.columns:
    medals_corr = correlation_matrix["medals"]
    print("\nCorrelation with 'medals':")
    print(medals_corr)
else:
    print("Column 'medals' does not exist in the correlation matrix.")
```

```python
train = teams[teams["year"]<2012].copy()
test = teams[teams["year"]>=2012].copy()
```

```python
train
```

```python
test
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
predictors = ["athletes","prev_medals"]
target="medals"
```

```python
reg.fit(train[predictors],train["medals"])
```

```python
LinearRegression()
```

```python
predictions = reg.predict(test[predictors])
```

```python
test["predictions"]=predictions
```

```python
test
```

```python
test.loc[test["predictions"] < 0, "predictions"] = 0
```

```python
test["predictions"]=test["predictions"].round()
```

```python
test
```

```python
from sklearn.metrics import mean_absolute_error
error = mean_absolute_error(test["medals"],test["predictions"])
```
